```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_excel('C:\\Users\\admin\\OneDrive\\Desktop\\2ND SEM FILES\\ML\\Data File.xlsx')
df
```

```python
#Label encoding and One hot encoding
encoded_df = pd.get_dummies(df['Toss winner'], prefix='Toss winner')
df_encoded = pd.concat([encoded_df, df], axis=1)
df_encoded = df_encoded.drop(columns=['Toss winner'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_encoded = df_encoded.apply(label_encoder.fit_transform)

pd.DataFrame(df_encoded)
```

```python
x = df_encoded.iloc[:, :-1]
y = df_encoded.iloc[:, -1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
x_test
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(x_train,y_train)
 

y_pred= model.predict(x_test)
y_pred

#The predicted values by the model for the entire data
Y_pred=model.predict(x)

#The predicted data for the entire model
df_ypred = pd.DataFrame(Y_pred,columns = ['Prediction'])
result = pd.DataFrame(pd.concat ([df.iloc[:, :-1], df_encoded['Result'], df_ypred],axis = 1 ))
result

#The accuracy score
from sklearn.metrics import accuracy_score
print (accuracy_score (y_test,y_pred))
```
